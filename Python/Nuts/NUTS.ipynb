#Metropolitan

```python
a = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Agrimonia/Nuts/Metropolitan.xlsx')
a.rename(columns = {'NUTS ID':'NUTS_ID'}, inplace = True)

a.head()  
```

# Mountain regions

```python
b = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Agrimonia/Nuts/Mountain regions.xlsx')
b.head()
```

# Urban-rural

```python
c = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Agrimonia/Nuts/Urban-rural.xlsx')
c.head()
```

# Join of the three tables

```python
d = a.merge(b, how='inner', on='NUTS_ID')
print(a.shape)
print(b.shape)
print(d.shape)
d.head()
```

```python
e = a.merge(c, how='inner', on='NUTS_ID')
print(a.shape)
print(c.shape)
print(e.shape)
e.head()
```

```python
f = d.merge(e, how='inner', on='NUTS_ID')
print(d.shape)
print(e.shape)
print(f.shape)
f.head()
```

# Save the result

```python
f.to_csv('/content/drive/MyDrive/Colab Notebooks/Agrimonia/Nuts/resultNuts3.csv')
```
