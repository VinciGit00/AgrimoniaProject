## Libraries

```python
import pandas as pd
```

## Agrimonia dataset

```python

data = pd.read_csv('https://zenodo.org/record/6620530/files/Agrimonia_Dataset.csv?download=1')
pd.set_option('display.max_columns', None)
data.head()
```

```python
shape = data.shape
  
# Printing Number of columns
print('Number of rows: ' +str(shape[0])+ ", number of columns: "+str(shape[1]))
```

```python
data.describe()
```
